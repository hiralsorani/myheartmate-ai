```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# df = pd.read_csv("cardio_train.csv", sep=';')
```

```python
df = pd.read_csv("cardio_train.csv", sep=';')
df
```

# 2. Remove duplicate rows

```python
df = df.drop(columns=['id'])

df.drop_duplicates(inplace=True)
print(df.shape)

```

## 1. Convert age from days to years


```python

df['age_years'] = (df['age'] / 365).astype(int)

print(df[['age','age_years']].head())
df.drop(columns=['age'], inplace=True)
```

# 3. if Nan Numerical columns → Fill with mean
# Categorical columns → Fill with mode

```python
# Numerical
df.fillna(df.mean(numeric_only=True), inplace=True)

# Categorical
for col in df.select_dtypes(include='object'):
        df[col].fillna(df[col].mode()[0], inplace=True)
df
```

# Check NaNs after replacement

```python
df.isna().sum()
```

# 5. Remove rows where diastolic BP > systolic BP


```python

df1 = df[df['ap_lo'] > df['ap_hi']]
df1
```

```python
df = df[df['ap_lo'] <= df['ap_hi']]
df
```

# 6. Create a new feature: BMI =  weight/height ratio (Body Mass Index)

```python
df['BMI'] = df['weight'] / ((df['height']/100) ** 2)
print(df[['weight', 'height', 'BMI']].head(10))
```

# 8. Create a new feature: pulse pressure

```python
df['pulse_pressure'] = df['ap_hi'] - df['ap_lo']
print(df[['ap_hi', 'ap_lo', 'pulse_pressure']].head(10))
```

## Exploratory Data Analysis (EDA)

# 11. Correlation Heatmap of whole data set

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

# 12. Outlier Detection using Boxplots

```python
num_cols = ['age_years', 'height', 'weight', 'ap_hi', 'ap_lo']
plt.figure(figsize=(15,5))
for i, col in enumerate(num_cols, 1):
    plt.subplot(1, len(num_cols), i)
    sns.boxplot(y=df[col])
    plt.title(f"{col} Boxplot")
plt.tight_layout()
plt.show()
```

# 13. Feature Distribution with KDE(Kernel Density Estimate)


# It shows the probability density of a feature, giving a smooth curve instead of a histogram’s discrete bins.
age_years – Slightly right-skewed, most people are between ~45–60 years.

height – Narrow and symmetric, most heights are around 160–180 cm.

weight – Right-skewed, majority of weights are below 100, with a few higher values.

ap_hi (systolic BP) – Highly skewed with extreme outliers, most values are very low but some are extremely high.

ap_lo (diastolic BP) – Peaked distribution around normal range (~70–90) with some irregular spikes.

```python
plt.figure(figsize=(12,6))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.kdeplot(df[col], fill=True, color=np.random.rand(3,))
    plt.title(f"{col} Distribution")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Define outlier conditions for each column
outlier_conditions = {
    'age_years': (df['age_years'] < 40) | (df['age_years'] > 60),
    'height': (df['height'] < 140) | (df['height'] > 200),
    'weight': (df['weight'] < 40) | (df['weight'] > 130),
    'ap_hi': (df['ap_hi'] < 50) | (df['ap_hi'] > 200),
    'ap_lo': (df['ap_lo'] < 40) | (df['ap_lo'] > 120)
}

# Count outliers for each column
outlier_counts = {col: cond.sum() for col, cond in outlier_conditions.items()}

# Display number of outliers and the rows
# print(f"Number of outlier rows: {len(outlier_rows)}")
# print(outlier_rows)

# Display the counts
print("Outlier counts per column:")
for col, count in outlier_counts.items():
    print(f"{col}: {count}")
```
